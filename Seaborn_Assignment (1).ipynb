##Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Scatter Plot:
Use: Scatter plots are used to visualize the relationship between two numerical variables. They are helpful in identifying patterns, clusters, or correlations between variables.

Histogram:
Use: Histograms display the distribution of a single numerical variable. They are useful for understanding the underlying frequency distribution, identifying outliers, and assessing data skewness.

Bar Plot:
Use: Bar plots are used to compare the mean or aggregate values of different categories. They are suitable for showing categorical data where each category represents a distinct group or class.

Box Plot:
Use: Box plots, also known as box-and-whisker plots, are used to display the distribution of numerical data based on quartiles. They provide insights into the central tendency, variability, and presence of outliers within a dataset.

Heatmap:
Use: Heatmaps are effective for visualizing the correlation matrix between multiple variables. They are particularly useful in exploratory data analysis and identifying patterns of association or dependency between variables.

##Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
frmri_dataset = sns.load_dataset('fmri')
frmri_dataset
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data = frmri_dataset, x = 'timepoint', y = 'signal',hue ='event', style = 'region',markers= True)
plt.ylabel ('signal')
plt.xlabel ('Timepoint')
plt.title('FMRI Signal Over Time by Event and Region')
```

##Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x ='pclass', y = 'age' and y = 'fare'.

```python
data_titanic = sns.load_dataset('titanic')
data_titanic
```

```python

sns.set(style="whitegrid")

# Box plot for 'age'
sns.boxplot(x='pclass', y='age', data=data_titanic)
plt.title('Box plot of Age by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()

# Box plot for 'fare'
sns.boxplot(x='pclass', y='fare',data=data_titanic)
plt.title('Box plot of Fare by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

##Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
# Load the diamonds dataset
diamonds_data = sns.load_dataset('diamonds')

# Plot histogram with hue for 'cut'
sns.histplot(data=diamonds_data, x='price', hue='cut', bins=30, kde=True)
plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

##Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
# Load the iris dataset
iris_data = sns.load_dataset('iris')

# Plot pair plot with hue for 'species'
sns.pairplot(iris_data, hue='species')
plt.title('Pair Plot of Iris Dataset by Species')
plt.show()
```

##Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights_data = sns.load_dataset('flights')

# Reshape the dataset into a pivot table format
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot, cmap='coolwarm', annot=True, fmt="d")
plt.title('Passenger Count by Month and Year')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```
