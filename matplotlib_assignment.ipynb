##Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Matplotlib is a comprehensive library in Python used for creating static, animated, and interactive visualizations in a variety of formats. It is widely used for generating plots, histograms, power spectra, bar charts, error charts, scatterplots, etc. It provides a MATLAB-like interface through its pyplot module, making it easy to create plots with just a few lines of code.

Here are five types of plots that can be easily created using the Pyplot module of Matplotlib:

Line Plot: Line plots are used to visualize the relationship between two continuous variables. They are commonly used to display trends over time.

Histogram: Histograms are used to represent the distribution of a continuous variable. They display the frequencies of observations within certain ranges or bins.

Bar Plot: Bar plots are used to compare the values of different categories or groups. They display categorical data with rectangular bars, where the length or height of each bar represents the value.

Scatter Plot: Scatter plots are used to visualize the relationship between two variables. Each point on the plot represents a single observation, with the x-coordinate representing one variable and the y-coordinate representing another.

Pie Chart: Pie charts are used to represent the proportions of different categories within a dataset. They are circular and divided into slices, where each slice represents a category and the size of the slice represents the proportion of that category relative to the whole.

These are just a few examples, but Matplotlib's capabilities extend far beyond these basic plot types, allowing for the creation of complex and customized visualizations to suit various needs.








#Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
plt.scatter(x,y, color = 'blue', alpha = 0.5)
plt.title('scatter plot x against y')
plt.xlabel('X VARIABLE')
plt.ylabel('Y VARIABLE')
plt.grid(True)
plt.show()
```

##Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

```python
#line1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
#line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
#line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
#line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

The subplot() function in Matplotlib is used to create multiple plots within the same figure. It allows you to arrange plots in a grid layout, specifying the number of rows and columns, and the position of each plot within that grid. This function is particularly useful when you want to compare multiple plots or visualize related data side by side.

```python
# create subplot
plt.figure(figsize=(12,8))

#subplot1
plt.subplot(2,2,1)
plt.plot(x1,y1,color='red')
plt.title('Line 1')
plt.xlabel('x1')
plt.ylabel('y1')

#subplot2
plt.subplot(2,2,2)
plt.plot(x2,y2,color='orange')
plt.title('Line 2')
plt.xlabel('x2')
plt.ylabel('y2')

#subplot3
plt.subplot(2,2,3)
plt.plot(x3,y3,color='blue')
plt.title('Line 3')
plt.xlabel('x3')
plt.ylabel('y3')

#subplot4
plt.subplot(2,2,4)
plt.plot(x4,y4,color='purple')
plt.title('Line 4')
plt.xlabel('x4')
plt.ylabel('y4')
```

##Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.


```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```


A bar plot is a type of plot that presents categorical data with rectangular bars, where the length or height of each bar represents the value of the corresponding category. Bar plots are used to compare the values of different categories or groups, making it easy to visualize and interpret the relationships between them.

Comparison: Bar plots allow for a quick visual comparison of different categories or groups.

Trends: They help in identifying trends or patterns across different categories.

Frequency distribution: Bar plots can display the frequency distribution of categorical variables.

Relative proportions: They provide insight into the relative proportions of different categories within a dataset.

```python
plt.figure(figsize=(8,6))
plt.bar(company,profit,color='blue')
plt.title('Yearly profit comparison of top companies in cr')
plt.xlabel('Companies')
plt.ylabel('Profit')
```

#Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset based on five summary statistics: minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. It provides a concise way to visualize the central tendency, dispersion, and skewness of the data, as well as the presence of outliers.

Box plots are used for the following purposes:

Visualizing distribution: Box plots provide a clear summary of the distribution of the dataset, including its spread and skewness.
Identifying outliers: They help in identifying outliers, which are data points that significantly differ from the rest of the dataset.
Comparing distributions: Box plots enable easy comparison of the distributions of multiple datasets or groups.

```python
plt.figure(figsize=(8,6))
plt.boxplot([box1,box2],labels=['Box 1','Box 2'])
plt.title('Box plot comparison')
plt.xlabel('Box')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```
